```python
x_train, x_test, y_train, y_test = train_test_split(features, label, test_size=0.3,shuffle=True)

# Tạo model
model = LogisticRegression(C=0.7, solver='lbfgs', max_iter=1000)

# train model
model.fit(x_train, y_train)

# dự đoán
y_pred = model.predict(x_test)
```

```python
# Create an instance of tkinter frame or window
win=Tk()

# Set the size of the tkinter window
win.geometry("700x350")

def Diabetes():
    dict_columns = {
    'Pregnancies': [int(a.get())],
    'Glucose': [int(b.get())],
    'BloodPressure': [int(c.get())],
    'SkinThickness': [int(d.get())],
    'Insulin': [int(e.get())],
    'BMI': [float(f.get())],
    'DiabetesPedigreeFunction': [float(g.get())],
    'Age': [int(h.get())]
    }
    dataFrame = pd.DataFrame(dict_columns)
    result = model.predict(dataFrame)
    if result[0] == 1: messagebox.showwarning("Kết quả", "Bạn có nguy cơ cao bị tiểu đường")
    else: messagebox.showinfo("Kết quả", "Bạn ít có khả năng bị tiểu đường")

# Create an Entry widget
Label(win, text="Ứng dụng cảnh báo bệnh tiểu đường", font=('Calibir 18')).pack(pady=10)

Label(win, text="Xin mời bạn nhập các chỉ số:", font=('Calibir 10 underline')).place(x=300,y=50)

Label(win, text="Pregnancies", font=('Calibri 10')).place(x=200,y=75)
a=Entry(win, width=35)
a.place(x=300, y=75)
Label(win, text="Glucose", font=('Calibri 10')).place(x=200,y=100)
b=Entry(win, width=35)
b.place(x=300, y=100)
Label(win, text="BloodPressure", font=('Calibri 10')).place(x=200,y=125)
c=Entry(win, width=35)
c.place(x=300, y=125)
Label(win, text="SkinThickness", font=('Calibri 10')).place(x=200,y=150)
d=Entry(win, width=35)
d.place(x=300, y=150)
Label(win, text="Insulin", font=('Calibri 10')).place(x=200,y=175)
e=Entry(win, width=35)
e.place(x=300, y=175)
Label(win, text="BMI", font=('Calibri 10')).place(x=200,y=200)
f=Entry(win, width=35)
f.place(x=300, y=200)
Label(win, text="DiabetesPedigreeFunction", font=('Calibri 10')).place(x=200,y=225)
g=Entry(win, width=35)
g.place(x=300, y=225)
Label(win, text="Age", font=('Calibri 10')).place(x=200,y=250)
h=Entry(win, width=35)
h.place(x=300, y=250)


Button(win, text="Kiểm tra", height=2, command=Diabetes).place(x=350,y=275)

win.mainloop()
```
